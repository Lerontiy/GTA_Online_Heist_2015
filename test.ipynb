```python
with open("порядок діалогів оригінал.txt", encoding = 'utf-8', mode="r") as oryginal_file:
    oryginal_list = list(line for line in oryginal_file)

with open("порядок діалогів українською.txt", encoding = 'utf-8', mode="r") as ukrainskyi_file:
    ukrainskyi_list = list(line for line in ukrainskyi_file)

```

```python
filepath = "GTA_Online_Heist_2015/"
path = os.listdir(filepath)

i = 0
for content in path:
    if i>=5:
        break
    print(content)
    i+=1


```

```python
for oryginal_index in range(len(oryginal_list)):
    for filewrite_name in path:
        with open(f"{filepath}{filewrite_name}", encoding = 'utf-8', mode="r") as filewrite_open:
            filewrite_data = filewrite_open.readlines()

            for filewrite_index in range(len(filewrite_data)):
                filewrite_data[filewrite_index] = filewrite_data[filewrite_index].replace(oryginal_list[oryginal_index], ukrainskyi_list[oryginal_index])

        with open(f"{filepath}{filewrite_name}", encoding = 'utf-8', mode="w") as filewrite_open:
            filewrite_open.write("".join(filewrite_data))


```
